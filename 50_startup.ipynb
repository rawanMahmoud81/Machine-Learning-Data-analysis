```python
#Importing the libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, VotingRegressor, AdaBoostRegressor
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
from sklearn.svm import SVR
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import os
%matplotlib inline
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
#Get the dataset
df = pd.read_csv("/content/50_Startups.csv")
df.head()
```

```python
# prompt: Develop an ensemble regression model for the startups data set using Linear Regression, SVR., Random Forest,Â Adaboost

# Encoding categorical data
labelencoder = LabelEncoder()
df['State'] = labelencoder.fit_transform(df['State'])

# Define features (X) and target (y)
X = df.iloc[:, :-1].values
y = df.iloc[:, 4].values

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Initialize individual regressors
lr = LinearRegression()
svr = SVR()
rf = RandomForestRegressor()
ada = AdaBoostRegressor()


# Fit the models
lr.fit(X_train, y_train)
svr.fit(X_train, y_train)
rf.fit(X_train, y_train)
ada.fit(X_train, y_train)


# Make predictions
y_pred_lr = lr.predict(X_test)
y_pred_svr = svr.predict(X_test)
y_pred_rf = rf.predict(X_test)
y_pred_ada = ada.predict(X_test)


# Evaluate individual models
print("Linear Regression R-squared:", r2_score(y_test, y_pred_lr))
print("SVR R-squared:", r2_score(y_test, y_pred_svr))
print("Random Forest R-squared:", r2_score(y_test, y_pred_rf))
print("AdaBoost R-squared:", r2_score(y_test, y_pred_ada))


# Create an ensemble model using VotingRegressor
ensemble = VotingRegressor(estimators=[('lr', lr), ('svr', svr), ('rf', rf), ('ada', ada)])
ensemble.fit(X_train, y_train)
y_pred_ensemble = ensemble.predict(X_test)


# Evaluate the ensemble model
print("Ensemble R-squared:", r2_score(y_test, y_pred_ensemble))
```
